## Which player got the maximum premium on the base price?

```python
df["Premium"] = df["SOLD PRICE"]-df["BASE PRICE"]
df["Premium"]
```

```python
High_premium_player_name = df.loc[df['Premium'].idxmax()]
High_premium_player_name
```

## Which players got the maximum premium offering on their base price?

```python
max_premium = df["Premium"].max()
max_premium
```

```python
player_name = df[df["Premium"] == max_premium]
print(player_name['PLAYER NAME'])
```

## What is the average SOLD PRICE for each age category?


```python
average_sold_price = df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
average_sold_price.columns = ['AGE', 'Average SOLD PRICE']
print(average_sold_price)
```

## Average SOLD PRICE for Different Playing Roles in Each Age Category?

```python
average_sold_price = df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
average_sold_price.columns = ['AGE','PLAYING ROLE','Average SOLD PRICE']
print(average_sold_price)
```
