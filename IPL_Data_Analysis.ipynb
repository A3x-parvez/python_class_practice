Which player got the maximum premium on the base price?

```python
df['Premium'] = df['SOLD PRICE'] - df['BASE PRICE']

max_premium_player = df.loc[df['Premium'].idxmax()]

print(f"Player with maximum premium: {max_premium_player['PLAYER NAME']} with a premium of {max_premium_player['Premium']}")
```
